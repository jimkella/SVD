```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

from PIL import Image
image = Image.open('lake.png').convert('L')

# Compute the SVD of the image
U, s, V = np.linalg.svd(image)

# Reconstruct the image using the first k singular values
k = 100  # choose the number of singular values to use
reconstructed_image = U[:, :k] @ np.diag(s[:k]) @ V[:k, :]

# Display the original and reconstructed images
plt.imshow(image, cmap='gray')
plt.show()
plt.imshow(reconstructed_image, cmap='gray')
plt.show()

#im = Image.fromarray(reconstructed_image)
#im.save("your_file.png")
matplotlib.image.imsave('lake_gray.png', image,cmap='gray')
matplotlib.image.imsave('lake_gray_SVD.png', reconstructed_image,cmap='gray')
```
